```python
import os
import glob
import json
from pymongo import MongoClient
```

```python
client = MongoClient('localhost',27017)
db = client['PrenatalTracker']
path = "C:\\Users\\nitar\\Desktop\\3.27.23\\JSONData"
```

```python
# loop through all JSON files in folder
for filename in glob.glob(os.path.join(path, '*.json')):
    # read JSON file
    with open(filename, 'r') as f:
        json_data = json.load(f)
    # extract collection name from filename
    collection_name = os.path.splitext(os.path.basename(filename))[0]
    # insert JSON data into MongoDB collection
    collection = db[collection_name]
    collection.insert_many(json_data)
```

```python
# loop through collections in database
for collection_name in db.list_collection_names():
    # get the collection
    collection = db[collection_name]
    # update the field name using $rename
    collection.update_many({}, {'$rename': {'ABP': 'BP'}})
```

```python
collection_names = db.list_collection_names()
for name in collection_names:
    print(f'{name} = "{name}"')
```
